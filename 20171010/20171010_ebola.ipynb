# Extract the cases and deaths in Liberia

Extract the cases and deaths in Liberia from the ebola dataset.

Change the column names to date, cases, and deaths.

Add a column called ratio that has the ratio of deaths to cases for each observation.

```python
ebola_liberia = ebola_df[["Date", "Cases_Liberia", "Deaths_Liberia"]]
ebola_liberia = ebola_liberia.rename(columns={"Date": "date", "Cases_Liberia": "cases", "Deaths_Liberia": "deaths"})
ebola_liberia["ratio_death_to_cases"] = ebola_liberia["deaths"]/ebola_liberia["cases"]

ebola_liberia.head(10)
```

This data has earliest dates last and latest dates first. Often, we want our data in chronological order. Change the dataset so it's in chronological order, from the observation with the earliest date to the one with the latest.

Filter out all rows from the ebola_liberia dataframe that are missing death counts for Liberia. How many rows are in the dataframe now?

```python
ebola_liberia = ebola_liberia.sort_values(by="date")

deaths_col_index = ebola_liberia.columns.get_loc("deaths")
ebola_liberia = ebola_liberia.dropna(thresh=deaths_col_index)
rows, columns = ebola_liberia.shape
print("The number of rows after filtering out NaNs is", rows)
```


Create a new object called first_five that has only the five observations with the highest death counts in Liberia. What date in this dataset had the most deaths? date.

```python
first_five = ebola_liberia.nlargest(5, columns=("deaths"))
max_datetime = first_five.nlargest(1, columns=("deaths"))["date"].values[0]
print(max_datetime)

# Pretty print the date.
print(pd.to_datetime(max_datetime).strftime('%Y-%m-%d'))
```
