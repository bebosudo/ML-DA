```python
import pandas as pd
# import request

# The first column was unnamed, so we overwrite the header names.
col_names = ['name', 'team', 'position', 'time', 'shots', 'passes', 'tackles', 'saves']
world_cup_df = pd.read_csv("https://raw.githubusercontent.com/ferdas/faraway_csv/d2ac718a8137ba238/datasets/worldcup.csv", names=col_names, header=1)

# Search for players that did at least one save for Italy or Denmark teams.
world_cup_df[(world_cup_df["saves"] != 0) & ((world_cup_df["team"] == "Italy") | (world_cup_df["team"] == "Denmark"))]
```

The top10 players with the largest number of passes:

```python
world_cup_df.nlargest(10, columns=("passes"))
```

```python
import matplotlib.pyplot as plt
plt.plot(world_cup_df["name"], world_cup_df["passes"],  'o')
plt.show()
```
